# Obtain information about directories for visualization.

File 1:
    One column with all nodes. Extended rows with all children of parent nodes

File 2:
    Columns: File name, size, extension. All files and directories. Directories have empty size and extension values.

```python
import os

os.chdir('/home/ubuntu/IBD-Network-Models-Reproduction/')
startpath = 'PAPER_DATA'
```

```python
# Function to print directory with indents
def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))
            
list_files(startpath)
```

```python
# File 1: Make dictionary of parents as keys with values of list of children
directories = {}
for root, dirs, files in os.walk(startpath):
    parent = root # The root is the parent node
    children = [root+'/'+x for x in (dirs + files)] # All files and directories are children
    directories[parent] = children # Add to dictionary: Parent = key, Children = values
    directories.update(dict(zip([root+'/'+x for x in files], [[]]*len(files)))) 
    # Add files as nodes with no children
```

```python
# Write file 1 to csv
import csv
filename = 'DirectoryNodes_wPaths.csv'

with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for key in directories:
        writer.writerow([key] + directories[key])
```

```python
# Compile name, size, and extension for file 2
path = []
name = []
size = []
ext = []
for root, dirs, files in os.walk(startpath):
    # Add root as directory (ignore dirs, as they will all be roots)
    path.append(root)
    name.append(os.path.basename(root))
    size.append(None)
    ext.append(None)
    # Add files individually
    for file in files:
        filename, file_extension = os.path.splitext(file)
        file = root +'/'+ file
        path.append(file)
        name.append(filename)
        size.append(os.stat(file).st_size)
        ext.append(file_extension)

# Export file2 to csv with pandas
import pandas as pd
nodes = pd.DataFrame({'Path':path, 'Name':name, 'Size':size, 'Type':ext})
nodes.to_csv('NodeTypes_wPaths.csv', index=False)
```
